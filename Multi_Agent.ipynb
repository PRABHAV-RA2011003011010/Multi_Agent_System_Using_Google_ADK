```python
try:
    from google.adk.agents import Agent
    from google.adk.runners import InMemoryRunner
    from google.adk.tools import google_search
    from google.genai import types

    print("✅ ADK components imported successfully.")
except ModuleNotFoundError as e:
    # Friendly guidance for missing dependencies
    print("ModuleNotFoundError:", e)
    print()
    print("It looks like the Google ADK/GenAI packages are not installed in this Python environment.")
    print("To install the typical required packages, run the following in PowerShell:")
    print("pip install --upgrade google-genai google-adk python-dotenv")
    print()
    print("Or create and activate a virtual environment, then install: ")
    print("python -m venv .venv; .venv\Scripts\Activate; pip install -r requirements.txt")
    # Re-raise so the notebook indicates failure if the user wants that behavior
    raise
```

```python
root_agent = Agent(
    name="helpful_assistant",
    model="gemini-2.5-flash-lite",
    description="A simple agent that can answer general questions.",
    instruction="You are a helpful assistant. Use Google Search for current info or if unsure.",
    tools=[google_search],
)

print("✅ Root Agent defined.")
```

```python
import os
from google import generativeai as genai
from dotenv import load_dotenv

# Load environment variables
load_dotenv()

# Get API key from environment
api_key = os.getenv('GEMINI_API_KEY')
if not api_key:
    raise ValueError("GEMINI_API_KEY not found in environment variables")

# Configure the Gemini API
genai.configure(api_key=api_key)

print("✅ API key loaded and configured successfully.")
```

```python
# Create a runner to manage the agent
runner = InMemoryRunner(agent=root_agent, app_name="TestApp")

# Run the agent with a test message
response = runner.run(
    message="What is artificial intelligence?",
)

# Print the response
print("Agent's response:")
print(response.text)
```
